```python
qualified = []

holding = words[:8]

index = 8

while index < len(words):
    
    if holding_qualifies(holding):
        qualified.append(holding)
    else:
        holding.pop(0)
        holding.append(words[index])
    
    index += 1
    
    
def holding_qualifies(holding):
    # if it contains lowercaseUPPERCASE
    first = "".join(holding[:4])
    second = "".join(holding[4:])
    
    if first.isLower() and second.isUpper():
        return first == second.lower()
    return False
```
